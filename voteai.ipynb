```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/HP/Downloads/archive (17)/Vote Ai.csv')
```

```python
 df.isnull().sum()[df.isnull().sum() >0]
```

```python
#fill numerical data with median
num_cols = ['Assets_Cr', 'Liabilities_Cr', 'Campaign_Spending_Cr', 'Fake_News_Score' , 'Deepfake_Propagation_Score', 'Vote_Surge_Score','Popularity_Score', 'Historical_Win_Rate', 'Registered_Voters', 'EVM_Votes', 'Time_Between_Updates_min', 'Adversarial_Pressure', 'Influence_Network_Strength', 'Vote_Change_Per_Round', 'Server_Delay_ms', 'Noise_Float_1', 'Noise_Float_2']
for cols in num_cols:
    df[cols] = df[cols].fillna(df[cols].median())
```

```python
#for categorical columns
cat_cols = ['Speech_Sentiment', 'Attack_Type', 'Defense_Strategy']
for cols in cat_cols:
    df[cols] = df[cols].fillna('Unknown')
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df.to_csv('vote_ai_clean.csv', index=False)
```
